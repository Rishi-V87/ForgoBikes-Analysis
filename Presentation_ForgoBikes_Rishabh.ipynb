```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
## Importing dataset for individual 'Months' for 2018 and 2019:

bike_2018 = pd.read_csv('bike_2018')
bike_2019 = pd.read_csv('bike_2019')
```

# Distribution of Customers by Gender:

* There are **More Male** customers than **Female** customers for the year 2018 and 2019.
* There were approx. **~300,000** more Males as compared to **~85,000** more Females in 2019 riding / renting bikes.

This can be a great **Marketing opportunity** to promote the services to **Men** segment.

```python
def plot_bars_2(df, x_col, color, xlabel, ylabel, title):
    sb.countplot(x= x_col, data = df, color = color)
    plt.xlabel(xlabel, fontsize = 14)
    plt.ylabel(ylabel, fontsize = 14)
    plt.title(title, fontsize = 18, fontweight="bold")

fig = plt.figure(figsize = [15, 4])

## 2018
plt.subplot(1, 2, 1)
plot_bars_2(bike_2018, 'member_gender', 'gold', 'Gender','Number of Customers', 'Gender Distribution - 2018')

## 2019
plt.subplot(1, 2, 2)
plot_bars_2(bike_2019, 'member_gender', 'skyblue', 'Gender', 'Number of Customers', 'Gender Distribution - 2019')
```

## Distribution of Duration (Mins) by Age:

* Younger people are more likely to travel a greater distance, upto 1400 mins, as compared to older people. In both the years, this age group happens to be **Less than 40 years**
* Also, there are **more people** who rented the bikes in 2019 as compared to 2018. This is visible from the color density of the graphs.


The given analysis provides an opportunity to promote the **Bike Renting** to the **Younger Population** who are not only **more likely to rent the bike** but also to  use it for a **longer duration**.

```python
fig = plt.figure(figsize = [20, 8])

# 2018
plt.subplot(1, 2, 1)
sb.scatterplot(x='Age', y = 'duration_minutes', data = bike_2018, alpha = 0.025, color = 'darkred')
plt.xlim(15, 90)
plt.xlabel( 'Age', fontsize = 14)
plt.ylabel( 'Duration (Mins)', fontsize = 14)
plt.title( 'Distribution - Duration v/s Age - 2018', fontsize = 18, fontweight="bold")

# 2019
plt.subplot(1, 2, 2)
sb.scatterplot(x='Age', y = 'duration_minutes', data = bike_2019, alpha = 0.025, color = 'blue')
plt.xlim(15, 90)
plt.xlabel( 'Age', fontsize = 14)
plt.ylabel( 'Duration (Mins)', fontsize = 14)
plt.title( 'Distribution - Duration v/s Age - 2019', fontsize = 18, fontweight="bold")
print(" ")
```

### Distribution of Place of Renting Bikes by Gender (2019):

* The proportion of number of **females** for the **Top 20 places of renting bikes** is very **constant** when compared to **Male** counterparts. It follows a **gradual decrease** as the number of customers decrease. 
* There seems to be no **Location preferences** from either **Men** or **Women**.

```python
## 2019
fig = plt.figure(figsize = [15, 15])
sb.countplot(y = 'start_station_name', data = bike_2019, palette='Purples_r', hue= 'member_gender',
             order = bike_2019['start_station_name'].value_counts().index, orient = 'v')
plt.ylim(-0.5,19.5)
plt.xlabel( 'Number of Customers', fontsize = 14)
plt.ylabel( 'Top Bikes Rental Places', fontsize = 14)
plt.title( 'Gender Distribution v/s Top Rental Places - 2019', fontsize = 18, fontweight="bold")
print(" ")
```

### Distribution of Duration (Mins) by Month and Gender:

* There seemed to be a clear difference in the **Average Duration of riding/renting bikes** among the **Male** and the **Female** for both the years.
* Also, there is a slight fluctuation in the **Mean Duration (Mins)** among the **Months** in the year **2019**. The **Average Duration** time for the month of **January and February** are pretty low for **Male and Female** when compared to the rest of the months for the year **2019**.


From the promotional point of view, this analysis sets a great tone to **reach out to Potential Female** customers and tell them about the benfits of riding bikes by using the examples of **existing Female Customer group**.



```python
fig = plt.figure(figsize = [20, 15])

## 2018
plt.subplot(2, 1, 1)
sb.pointplot(data = bike_2018, x = 'start_month', y = 'duration_minutes', hue = 'member_gender',
                  dodge = 0.3, linestyles = "")
plt.xlabel( 'Month', fontsize = 14)
plt.ylabel( 'Duration (Mins)', fontsize = 14)
plt.title( 'Distribution - Duration (Mins) v/s Month v/s Gender - 2018', fontsize = 18, fontweight="bold")


## 2019
plt.subplot(2, 1, 2)
sb.pointplot(data = bike_2019, x = 'start_month', y = 'duration_minutes', hue = 'member_gender',
                  dodge = 0.3, linestyles = "")
plt.xlabel( 'Month', fontsize = 14)
plt.ylabel( 'Duration (Mins)', fontsize = 14)
plt.title( 'Distribution - Duration (Mins) v/s Month v/s Gender - 2019', fontsize = 18, fontweight="bold")
print(" ")
```
